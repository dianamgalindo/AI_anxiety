```python
from bs4 import BeautifulSoup
import requests
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'}

url='https://www.google.com/search?q=ai+anxiety&amp;ie=utf-8&amp;oe=utf-8&amp;num=10'
html = requests.get(url,headers=headers)
soup = BeautifulSoup(html.text, features="html.parser")
```

```python
SearchData = soup.find_all("div", {"id": "search"})
SearchData
```

```python
g = 0
Links = []
l ={}

for i in range(0, len(SearchData)):
    link = SearchData[i].find('a').get('href')
    
    if(link is not None):
        if(link.find('https')!= -1 and link.find('hhtp')==0 and link.find('aclk')== -1):
            g = g+1
            Links.append(link)
    else:
        continue
            
print(Links)
```
